## Definiciones clave

- **Match:** palabra correcta.
- **Substitution:** palabra A fue reemplazada por palabra B.
- **Deletion:** falta una palabra de la referencia.
- **Insertion:** aparece una palabra que no estaba.
- **Alineación:** emparejamiento entre referencia e hypothesis para poder contar errores.

## Tipos de alineación

| Reference | Hypothesis | Tipo |
|---|---|---|
| igual | igual | match |
| palabra A | palabra B | substitution |
| palabra | None | deletion |
| None | palabra | insertion |


```python
pairs = [
    ("pedido", "perdido"),
    ("mañana", "mañana"),
    ("urgente", None),
    (None, "eh"),
    ("ocho", "dos"),
]

for ref, hyp in pairs:
    if ref == hyp:
        label = "match"
    elif ref is None:
        label = "insertion"
    elif hyp is None:
        label = "deletion"
    else:
        label = "substitution"
    print(f"ref={str(ref):<10} hyp={str(hyp):<10} -> {label}")
```

## Lectura docente

Este ejercicio prepara el terreno para WER. Antes de fórmula, necesitamos poder nombrar los errores.

**Pregunta:** ¿qué substitution es más peligrosa: `pedido -> perdido` o `ocho -> dos`?


```python
critical_pairs = [("ocho", "dos"), ("comprar", "vender"), ("mil", "millón")]

for ref, hyp in critical_pairs:
    print(f"{ref} -> {hyp}: revisar por dominio")
```
