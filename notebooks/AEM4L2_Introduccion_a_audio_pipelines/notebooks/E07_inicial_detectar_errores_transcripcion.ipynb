## Tipos de alineación

| Reference | Hypothesis | Tipo |
|---|---|---|
| igual | igual | match |
| palabra A | palabra B | substitution |
| palabra | None | deletion |
| None | palabra | insertion |


```python
pairs = [
    ("pedido", "perdido"),
    ("mañana", "mañana"),
    ("urgente", None),
    (None, "eh"),
    ("ocho", "dos"),
]

for ref, hyp in pairs:
    if ref == hyp:
        label = "match"
    elif ref is None:
        label = "insertion"
    elif hyp is None:
        label = "deletion"
    else:
        label = "substitution"
    print(f"ref={str(ref):<10} hyp={str(hyp):<10} -> {label}")
```
