# AEM4L2 E07 - Transformers vs Diffusion Models en audio

**Objetivo:** entender, sin matemática pesada, por qué usamos modelos tipo Transformer para transcribir/resumir y dónde entran los modelos de difusión.

Este notebook es teórico-visual y conecta la lecture con decisiones de arquitectura.

**Python exercise relacionado:** todos los scripts L2 usan ASR + LLM; este notebook explica el criterio de modelo.


## Dos familias de modelos

| Familia | Idea simple | Mejor para |
|---|---|---|
| Transformers | mirar contexto de una secuencia | ASR, traducción, resumen, clasificación |
| Diffusion models | empezar con ruido y refinar paso a paso | generación, restauración, síntesis de audio |

Frase docente: **Transformers interpretan secuencias; diffusion reconstruye o genera señales.**


## Transformers

Idea base:

> Un Transformer mira muchas partes de la secuencia al mismo tiempo para entender contexto.

Analogía:

Un intérprete no traduce palabra por palabra de forma ciega. Escucha la frase completa y usa contexto para resolver ambigüedad.

En audio/texto sirve para:

- transcribir;
- resumir;
- clasificar intención;
- traducir;
- extraer entidades.


````python
sequence = ["el", "pedido", "no", "llegó", "hoy"]
attention_hint = {
    "pedido": ["llegó"],
    "no": ["llegó"],
    "hoy": ["urgencia"],
}

for token in sequence:
    print(f'{token:<8} mira contexto -> {attention_hint.get(token, [])}')
````

## Diffusion models

Idea base:

> Un modelo de difusión empieza con ruido y lo va refinando hasta producir una señal útil.

Analogía:

Como restaurar una foto borrosa: paso a paso se va limpiando la imagen o el sonido.

En audio sirve para:

- generar voz;
- restaurar audio;
- reducir ruido;
- crear música o efectos;
- text-to-speech.


````python
steps = [
    (0, 'ruido puro'),
    (1, 'forma general de voz'),
    (2, 'sílabas más claras'),
    (3, 'voz natural'),
]

for step, description in steps:
    print(f'paso {step}: {description}')
````

## Comparación práctica

| Necesidad | Elección razonable | Por qué |
|---|---|---|
| Transcribir llamada en tiempo real | Transformer/ASR | necesita secuencia + contexto + velocidad |
| Resumir reunión | Transformer/LLM | necesita lenguaje y contexto |
| Restaurar audio con estática | Diffusion | reconstruye señal con detalle |
| Generar voz natural | Diffusion/TTS moderno | produce audio de alta fidelidad |
| Clasificar intención | Transformer/LLM | interpreta significado |


## Mini-caso de arquitectura

Call center necesita:

1. transcribir llamadas en vivo;
2. resumir reclamos;
3. restaurar partes con estática.

Diseño posible:

```text
WAV -> Transformer ASR -> transcript -> Transformer LLM -> summary
              │
              └── si audio dañado: módulo de restauración/diffusion antes o en paralelo
```

No todo lo resuelve un solo modelo.


````python
requirements = [
    ('transcripción rápida', 'Transformer ASR'),
    ('resumen accionable', 'Transformer LLM'),
    ('restauración de audio', 'Diffusion / denoising'),
    ('gate de calidad', 'métricas + reglas'),
]

for need, tool in requirements:
    print(f'{need:<24} -> {tool}')
````
