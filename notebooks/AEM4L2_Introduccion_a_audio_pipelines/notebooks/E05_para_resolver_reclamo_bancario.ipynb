## Fórmula

```text
WER = (S + D + I) / N
```

| Símbolo | Significado |
|---|---|
| S | substitutions |
| D | deletions |
| I | insertions |
| N | cantidad de palabras en la referencia |

WER responde: **qué proporción de palabras salió mal**.

WER no responde: **qué tan grave fue el error**.


````python
reference = "necesito cancelar mi turno medico de mañana"
hypothesis = "necesito cambiar mi turno medico mañana"

print('REF:', reference)
print('HYP:', hypothesis)
````

## Tabla manual

| Posición | Referencia | Hypothesis | Resultado |
|---:|---|---|---|
| 1 | necesito | necesito | match |
| 2 | cancelar | cambiar | substitution |
| 3 | mi | mi | match |
| 4 | turno | turno | match |
| 5 | medico | medico | match |
| 6 | de | - | deletion |
| 7 | mañana | mañana | match |

Entonces:

```text
S = 1
D = 1
I = 0
N = 7
WER = (1 + 1 + 0) / 7 = 28.57%
```


````python
S = 1
D = 1
I = 0
N = len(reference.split())
WER = (S + D + I) / N
print(f'S={S}, D={D}, I={I}, N={N}')
print(f'WER={WER:.2%}')
````

## Visualización simple

Una barra ayuda a explicar WER sin fórmulas abstractas.


````python
error_rate = WER
filled = round(error_rate * 30)
bar = '█' * filled + '-' * (30 - filled)
print(f'WER [{bar}] {error_rate:.2%}')
````

## WER bajo, riesgo alto

Ahora miremos otro caso:

```text
Referencia: tomar medicación cada ocho horas
Hypothesis: tomar medicación cada dos horas
```

Puede ser un WER bajo, pero el cambio `ocho -> dos` es crítico.


````python
reference_med = "tomar medicacion cada ocho horas"
hypothesis_med = "tomar medicacion cada dos horas"
S = 1
D = 0
I = 0
N = len(reference_med.split())
wer_med = (S + D + I) / N
print(f'WER médico={wer_med:.2%}')
print('Pero el error cambia la frecuencia de medicación -> riesgo crítico')
````
