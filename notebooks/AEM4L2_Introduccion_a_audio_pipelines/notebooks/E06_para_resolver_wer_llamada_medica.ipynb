## Transcript crudo

Un transcript es evidencia, pero obliga a leer todo.

```text
Hola, llamo porque el pedido 4521 no llegó. Lo necesitaba para hoy.
Quiero saber si pueden enviarlo urgente o cancelar la compra.
```

Para una persona es entendible. Para un backend todavía es poco operativo.


````python
transcript = (
    "Hola, llamo porque el pedido 4521 no llegó. "
    "Lo necesitaba para hoy. Quiero saber si pueden enviarlo urgente "
    "o cancelar la compra."
)
print(transcript)
````

## Resumen libre

Un resumen libre ayuda a humanos:

```text
El cliente reclama que el pedido 4521 no llegó y necesita una solución urgente.
```

Pero el backend no puede confiar en que siempre venga igual.

Problemas:

- no hay campos fijos;
- puede cambiar el formato;
- no hay valores cerrados;
- es difícil filtrar por urgencia o intent.


````python
free_summary = "El cliente reclama que el pedido 4521 no llegó y necesita una solución urgente."
print(free_summary)
````

## JSON mínimo

Ahora convertimos el mismo caso en campos.

| Campo | Para qué sirve |
|---|---|
| summary | lectura rápida humana |
| urgency | priorización |
| action_items | próximas acciones |

Este es el salto de texto lindo a dato útil.


````python
minimal = {
    "summary": "El cliente reclama que el pedido 4521 no llegó.",
    "urgency": "high",
    "action_items": [
        "Verificar estado del pedido 4521",
        "Ofrecer reenvío urgente o cancelación",
    ],
}

for key, value in minimal.items():
    print(f'{key}: {value}')
````

## Estructura completa

En soporte real podríamos agregar:

| Campo | Motivo |
|---|---|
| transcript | conservar evidencia |
| summary | lectura humana |
| customer_intent | ruteo automático |
| urgency | prioridad |
| action_items | operación |
| risk_notes | alertas o compliance |

En Python, Pydantic permite validar que esos campos existan y tengan tipos correctos.


````python
complete = {
    "transcript": transcript,
    "summary": "El cliente reclama que el pedido 4521 no llegó y pide solución urgente.",
    "customer_intent": "delivery_claim",
    "urgency": "high",
    "action_items": ["Revisar tracking", "Contactar al cliente"],
    "risk_notes": "Cliente menciona urgencia para hoy",
}

print('Campos de la estructura completa:')
for key in complete:
    print('-', key)
````

## Texto libre vs estructura validada

| Salida | Ventaja | Problema |
|---|---|---|
| Transcript | máxima evidencia | largo y sin decisión |
| Resumen libre | fácil de leer | formato variable |
| JSON mínimo | útil para backend | puede faltar contexto |
| Pydantic completo | validable y auditable | requiere diseñar schema |

Frase docente: **un buen schema convierte lenguaje en operación.**
