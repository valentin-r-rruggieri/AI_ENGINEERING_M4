## Definiciones clave

- **Error crítico:** error que puede causar daño o decisión incorrecta aunque el WER sea bajo.
- **Riesgo de dominio:** gravedad del error según contexto. Medicina, legal y finanzas suelen tener umbrales más estrictos.
- **Human review:** ruta de revisión humana cuando el sistema no debe automatizar.
