
| Tipo | Ejemplos | Herramienta inicial |
|---|---|---|
| CPU-bound | loops, regex pesada, JSON gigante | cProfile |
| I/O-bound | API, storage, DB, red | asyncio / batching |
| GPU-bound | inferencia pesada local | batching / hardware |
| Mixta | pipeline completo | medir y separar etapas |
