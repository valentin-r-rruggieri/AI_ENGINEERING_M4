# AEM4L5 E02 — Elegir despliegue

**Objetivo:** decidir entre serverless, servidor persistente o híbrido usando restricciones de negocio.


## Serverless vs servidor persistente

```mermaid
flowchart TD
    U["Usuario pide resumen"] --> D{"Patrón de tráfico"}
    D -->|"Bajo o irregular"| S["Serverless"]
    D -->|"Alto y estable"| P["Servidor persistente"]
    D -->|"Picos + SLA estricto"| H["Híbrido"]
    S --> S1["Paga por uso, posible cold start"]
    P --> P1["Modelo cargado, costo fijo"]
    H --> H1["Base caliente + escala bajo demanda"]
```

| Criterio | Serverless | Servidor persistente |
|---|---|---|
| Costo inactivo | Bajo o cero | Alto |
| Latencia inicial | Puede tener cold start | Baja |
| Tráfico ideal | Irregular | Alto y estable |
| Operación | Menos servidores | Más control |


## Matriz de decisión

La arquitectura correcta depende de latencia, costo, tráfico, tamaño del modelo y equipo disponible para operar.


````python
def elegir_despliegue(requests_dia: int, latencia_ms: int, trafico_irregular: bool, modelo_grande: bool) -> str:
    if latencia_ms <= 150 or (requests_dia > 20_000 and not trafico_irregular):
        return "Servidor persistente"
    if trafico_irregular and not modelo_grande:
        return "Serverless"
    if trafico_irregular and modelo_grande:
        return "Híbrido: servidor mínimo caliente + workers bajo demanda"
    return "Serverless o contenedor simple según costo"

escenarios = [
    ("resúmenes legales ocasionales", 30, 2000, True, False),
    ("chatbot tiempo real", 80_000, 100, False, True),
    ("campaña con picos impredecibles", 12_000, 1800, True, False),
    ("API interna estable", 10_000, 500, False, False),
]

for nombre, req, sla, irregular, grande in escenarios:
    print(f"{nombre:35} -> {elegir_despliegue(req, sla, irregular, grande)}")
````

## Lectura para clase

Serverless conviene cuando duele pagar servidores ociosos. Servidor persistente conviene cuando duele más el cold start que el costo fijo.
