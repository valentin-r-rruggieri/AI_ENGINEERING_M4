# AEM4L4 E07 — Tokenizacion: palabra, caracter y subword

**Objetivo:** Comparar tres niveles de tokenizacion y sus trade-offs.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

Word-level es simple pero falla con palabras desconocidas. Character-level cubre todo pero alarga mucho las secuencias. Subword divide palabras en piezas utiles y balancea cobertura con eficiencia.

## 2. Grafico guia

```mermaid
flowchart TB
    W[Palabra completa] --> W1[Menos tokens]
    W --> W2[Riesgo UNK]
    C[Caracter] --> C1[Cubre todo]
    C --> C2[Secuencias largas]
    S[Subword] --> S1[Balance cobertura/costo]
    S --> S2[Usado en LLMs modernos]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
word = "microfinanciamiento"
word_level = [word]
char_level = list(word)
subword = ["micro", "financ", "iamiento"]

for name, tokens in [("word", word_level), ("char", char_level), ("subword", subword)]:
    print(f"{name:<8} {len(tokens):>2} tokens -> {tokens}")
````

## 4. Interpretacion

Subword permite que una palabra rara conserve informacion semantica sin convertirse en una secuencia enorme de caracteres.

## 5. Errores comunes

- Presentar [UNK] como detalle menor.
- Olvidar que fragmentar demasiado sube latencia.
- Pensar que el tokenizer es independiente del dominio.
