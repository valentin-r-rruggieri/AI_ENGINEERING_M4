## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **Full fine-tuning** | Entrenar todos los parametros del modelo. |
| **PEFT** | Adaptar entrenando pocos parametros. |
| **LoRA** | Tecnica PEFT con matrices pequenas de bajo rango. |
| **Modelo congelado** | Base principal que no se modifica. |
| **Adapter** | Modulo pequeno para conocimiento especifico. |

## 3. Tres ejemplos guiados

Cada tema se baja a tres casos cortos. Para cada caso, preguntate: que cambia, que restriccion aparece y que decision tecnica se desprende.

| Ejemplo | Que muestra |
|---|---|
| **Un cliente** | Un adapter puede especializar sin tocar todo. |
| **Tres dominios** | Legal, medico y financiero comparten base. |
| **Actualizacion mensual** | Entrenar adapter puede ser mas barato. |


````python
base_gb = 14.0
full_por_cliente_gb = 14.0
lora_gb = 0.25
clientes = 8
full_total = base_gb + clientes * full_por_cliente_gb
lora_total = base_gb + clientes * lora_gb
print(f"Full fine-tuning: {full_total:.1f} GB")
print(f"Base + LoRA:      {lora_total:.1f} GB")
print(f"Ahorro aprox.:    {full_total - lora_total:.1f} GB")
````
