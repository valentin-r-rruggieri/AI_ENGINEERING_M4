## 1. Definiciones base

## 3. Tres ejemplos guiados

Cada tema se baja a tres casos cortos. Para cada caso, preguntate: que cambia, que restriccion aparece y que decision tecnica se desprende.

| Ejemplo | Que muestra |
|---|---|
| **Frase corta** | 'Revisar saldo' suele generar pocos tokens. |
| **Frase tecnica** | 'interchange fee capping' puede fragmentarse y requerir dominio. |
| **Texto largo** | Un contrato completo puede exigir recorte, resumen o RAG. |


````python
frases = [
    "Revisar saldo",
    "Necesito informacion sobre interchange fee capping",
    "El contrato establece confidencialidad y obligaciones postcontractuales",
]

def estimar_tokens(texto):
    palabras = texto.replace(",", " ,").split()
    extras = sum(1 for p in palabras if len(p) > 10)
    return len(palabras) + extras

for frase in frases:
    print(f"{estimar_tokens(frase):02d} tokens estimados -> {frase}")
````
