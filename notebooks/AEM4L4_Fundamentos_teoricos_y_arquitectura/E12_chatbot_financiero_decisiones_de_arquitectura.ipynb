## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **ADR** | Registro corto de una decision de arquitectura. |
| **SLA** | Objetivo operativo, por ejemplo responder en menos de 600 ms. |
| **Trade-off** | Costo de elegir una opcion frente a otra. |
| **Riesgo tecnico** | Falla o limite que puede afectar el sistema. |
| **Criterio** | Regla usada para elegir entre alternativas. |
