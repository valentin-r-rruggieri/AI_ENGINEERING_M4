# AEM4L4 E08 — BPE vs WordPiece

**Objetivo:** Comparar dos estrategias de subword tokenization.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

BPE fusiona pares frecuentes de simbolos. WordPiece usa piezas que mejoran la probabilidad del texto y suele marcar continuaciones con prefijos como ##.

## 2. Grafico guia

```mermaid
flowchart LR
    A[Corpus] --> B[BPE: fusiona pares frecuentes]
    A --> C[WordPiece: elige piezas probables]
    B --> D[financial + ization]
    C --> E[financial + ##ization]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
words = ["hypercholesterolemia", "financialization", "cat"]
examples = {
    "hypercholesterolemia": {"bpe": ["hyper", "cholesterol", "emia"], "wordpiece": ["hyper", "##cholesterol", "##emia"]},
    "financialization": {"bpe": ["financial", "ization"], "wordpiece": ["financial", "##ization"]},
    "cat": {"bpe": ["cat"], "wordpiece": ["cat"]},
}

for word in words:
    print(word)
    print("  BPE      ", examples[word]["bpe"])
    print("  WordPiece", examples[word]["wordpiece"])
````

## 4. Interpretacion

Ambas estrategias reducen [UNK], pero toman decisiones diferentes. La calidad depende del corpus, vocabulario y dominio.
