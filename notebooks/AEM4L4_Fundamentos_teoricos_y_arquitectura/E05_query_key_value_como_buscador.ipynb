## 3. Tres ejemplos guiados

````python
query = {"financiero", "credito", "clientes"}
keys = {
    "dinero": {"financiero", "pago"},
    "credito": {"financiero", "credito"},
    "plaza": {"espacio", "publico"},
    "clientes": {"personas", "financiero"},
}
for token, key in keys.items():
    print(f"{token:>8} -> coincidencias: {len(query & key)}")
````

## 6. Errores comunes y como corregirlos

| Error comun | Como corregirlo |
|---|---|
| **Q/K/V son lo mismo** | Tienen roles distintos. |
| **Value es etiqueta** | La etiqueta se parece mas a Key; Value es contenido. |
| **Solo importa token actual** | El token se reconstruye mirando otros tokens. |
