**Objetivo:** Construir un mapa simple donde cada token atiende a otros tokens relevantes.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

Self-attention calcula cuanto debe mirar cada token a los demas. El resultado se puede leer como una matriz: filas son tokens que buscan contexto, columnas son tokens que aportan informacion.

## 2. Grafico guia

```mermaid
    Prestamo --> Banco

```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
sentence = "El banco aprobo el prestamo despues de revisar el historial crediticio"
tokens = sentence.split()
attention = {token: {other: 0.05 for other in tokens} for token in tokens}
attention["banco"].update({"aprobo": 0.20, "prestamo": 0.35, "historial": 0.20, "crediticio": 0.15})

print(sentence)
print("Tokens que mas desambiguan 'banco':")
for token, weight in sorted(attention["banco"].items(), key=lambda item: item[1], reverse=True)[:5]:
    print(f"  {token:<12} peso={weight}")
````

## 4. Interpretacion

Banco se interpreta como institucion financiera porque los tokens financieros reciben mas peso contextual.

## 5. Errores comunes

- Confundir attention con explicacion perfecta del modelo.
- Asignar pesos sin relacion semantica.
- Olvidar que hay varias cabezas de attention en modelos reales.
