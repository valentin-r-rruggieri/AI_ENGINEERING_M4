## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **Modelo** | Componente que aprende patrones y produce predicciones o texto. |
| **Arquitectura** | Forma en que organizamos el sistema para procesar informacion. |
| **Token** | Unidad que el modelo procesa; puede ser palabra, parte de palabra o signo. |
| **Contexto** | Informacion disponible para interpretar una entrada. |
| **Latencia** | Tiempo que tarda el sistema en responder. |

## 3. Tres ejemplos guiados

Cada tema se baja a tres casos cortos. Para cada caso, preguntate: que cambia, que restriccion aparece y que decision tecnica se desprende.

| Ejemplo | Que muestra |
|---|---|
| **Chatbot lento** | Responde bien pero tarda 8 segundos: el problema es arquitectura, no solo calidad. |
| **Jerga financiera** | Terminos regulatorios nuevos pueden exigir adaptacion al dominio. |
| **Contrato largo** | Pegar todo el contrato aumenta tokens, costo y riesgo de ruido. |


````python
casos = [
    {"caso": "chatbot financiero", "tokens": 900, "latencia_ms": 780, "adaptar": True},
    {"caso": "faq interna", "tokens": 120, "latencia_ms": 180, "adaptar": False},
    {"caso": "resumen contrato", "tokens": 4200, "latencia_ms": 2600, "adaptar": True},
]

for c in casos:
    riesgos = []
    if c["tokens"] > 1000:
        riesgos.append("controlar contexto")
    if c["latencia_ms"] > 600:
        riesgos.append("optimizar latencia")
    if c["adaptar"]:
        riesgos.append("evaluar LoRA/PEFT")
    print(f"{c['caso']}: {', '.join(riesgos) or 'sin alerta principal'}")
````
