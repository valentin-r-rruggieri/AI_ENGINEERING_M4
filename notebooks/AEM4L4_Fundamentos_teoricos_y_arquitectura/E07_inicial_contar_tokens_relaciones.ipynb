# AEM4L4 E07 — Inicial | Contar tokens

**Objetivo:** Reforzar cálculo de relaciones.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Token count aproximado con split.

```python
phrases = ["El cliente pagó tarde", "El sistema bloqueó la cuenta por error"]

for phrase in phrases:
    tokens = phrase.split()
    n = len(tokens)
    print(phrase, "→", n, "tokens →", n*n, "relaciones")
```
