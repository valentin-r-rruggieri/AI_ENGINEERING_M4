# AEM4L4 E06 — Costo cuadratico de attention

**Objetivo:** Calcular por que duplicar tokens puede cuadruplicar relaciones de attention.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

En attention completa, cada token puede compararse con cada otro token. Por eso el numero aproximado de relaciones crece como N x N.

## 2. Grafico guia

```mermaid
flowchart LR
    A[50 tokens] --> B[2500 pares]
    C[100 tokens] --> D[10000 pares]
    E[Duplicar tokens] --> F[Cuadruplicar pares]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
for tokens in [10, 30, 50, 100, 500, 1000]:
    pares = tokens * tokens
    print(f"{tokens:>4} tokens -> {pares:>9} pares de attention")

base = 50
nuevo = 100
print("
Factor de crecimiento:", (nuevo * nuevo) // (base * base), "x")
````

## 4. Interpretacion

El contexto largo es potente, pero no gratuito. En produccion se controla con retrieval, resumen, chunking, cache y limites de input.

## 5. Errores comunes

- Medir latencia solo con ejemplos cortos.
- Creer que mas contexto siempre mejora calidad.
