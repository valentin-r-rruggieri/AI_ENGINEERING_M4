## 3. Tres ejemplos guiados

Cada tema se baja a tres casos cortos. La pregunta docente es: que cambia, que restriccion aparece y que decision tecnica se desprende.

| Ejemplo | Que muestra |
|---|---|
| **Banco financiero** | 'El banco aprobo el credito' activa dinero y credito. |
| **Banco asiento** | 'Me sente en el banco de la plaza' activa plaza y sentarse. |
| **Modelo** | Modelo en moda no significa lo mismo que modelo en IA. |


````python
pesos = {"el": 0.03, "banco": 0.15, "aprobo": 0.18, "credito": 0.34, "a": 0.02, "clientes": 0.28}
for token, peso in sorted(pesos.items(), key=lambda x: x[1], reverse=True):
    print(f"{token:>8} | {peso:.2f} {'#' * int(peso * 30)}")
````
