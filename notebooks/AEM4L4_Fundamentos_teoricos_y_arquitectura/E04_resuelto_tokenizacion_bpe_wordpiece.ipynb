# AEM4L4 E04 — Tokenización BPE vs WordPiece

**Objetivo:** Comparar estrategias de tokenización.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Subword tokenization equilibra flexibilidad y eficiencia.

```python
word = "microfinanciamiento"
tokenizations = {
    "word_level": ["microfinanciamiento"],
    "character_level": list(word),
    "bpe": ["micro", "financiamiento"],
    "wordpiece": ["micro", "##financia", "##miento"]
}

for name, tokens in tokenizations.items():
    print(name, "→", tokens)
```
