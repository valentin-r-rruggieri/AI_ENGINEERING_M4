# AEM4L4 E10 — PEFT vs full fine-tuning

**Objetivo:** Entender por que PEFT adapta menos parametros que full fine-tuning.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

Full fine-tuning actualiza todos o casi todos los parametros del modelo. PEFT pregunta si realmente hace falta entrenar todo, y agrega o selecciona un conjunto pequeno de parametros entrenables.

## 2. Grafico guia

```mermaid
flowchart LR
    B[Modelo base] --> F[Full fine-tuning]
    F --> C1[Copia completa cliente A]
    F --> C2[Copia completa cliente B]
    B --> P[PEFT]
    P --> A1[Adapter A]
    P --> A2[Adapter B]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
model_params = 7_000_000_000
peft_params = 70_000_000
print("Full FT entrena:", f"{model_params:,}", "parametros")
print("PEFT entrena   :", f"{peft_params:,}", "parametros")
print("Porcentaje PEFT:", round(peft_params / model_params * 100, 2), "%")
````

## 4. Interpretacion

PEFT baja costo, tiempo y storage, pero no elimina la necesidad de buenos datos ni evaluacion.

## 5. Errores comunes

- Decir que PEFT siempre mejora calidad.
- Ignorar privacidad y calidad de datos.
- No evaluar regresiones del comportamiento general.
