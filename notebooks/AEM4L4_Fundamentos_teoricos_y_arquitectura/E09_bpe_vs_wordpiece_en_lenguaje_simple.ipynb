## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **BPE** | Fusiona pares frecuentes para construir tokens. |
| **WordPiece** | Usa piezas de palabra, a veces con prefijo ##. |
| **Fusion** | Combinar piezas frecuentes en una mayor. |
| **Prefijo ##** | Marca que una pieza continua una palabra. |
| **Cobertura** | Representar palabras nuevas sin perder informacion. |
