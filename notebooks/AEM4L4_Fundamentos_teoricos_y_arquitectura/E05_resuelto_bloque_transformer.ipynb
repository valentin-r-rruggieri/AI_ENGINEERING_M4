## 1. Teoria breve

Un bloque Transformer no es solo attention. Attention mezcla contexto; la FFN transforma cada posicion; LayerNorm estabiliza; las conexiones residuales preservan informacion al apilar muchas capas.

## 2. Grafico guia

```mermaid
flowchart TB
    X[Entrada embeddings + posicion] --> A[Self-attention]
    X --> R1[Residual]
    A --> N1[Add + LayerNorm]
    R1 --> N1
    N1 --> F[Feed Forward Network]
    N1 --> R2[Residual]
    F --> N2[Add + LayerNorm]
    R2 --> N2
    N2 --> Y[Salida contextualizada]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
bloque = [
    ("self_attention", "mezcla informacion entre tokens"),
    ("add_norm_1", "estabiliza y conserva entrada"),
    ("ffn", "refina cada posicion por separado"),
    ("add_norm_2", "prepara salida para el siguiente bloque"),
]

representacion = "embedding inicial"
for paso, funcion in bloque:
    representacion = f"{paso}({representacion})"
    print(f"{paso:<16} -> {funcion}")

print("Salida conceptual:", representacion)
````

## 4. Interpretacion

Apilar bloques permite construir representaciones cada vez mas abstractas: primero relaciones locales, luego patrones de dominio y finalmente decisiones linguisticas.

## 5. Errores comunes

- Explicar Transformer como una sola capa.
- Omitir residuales y normalizacion.
- Confundir FFN con una red aplicada a toda la secuencia junta.
