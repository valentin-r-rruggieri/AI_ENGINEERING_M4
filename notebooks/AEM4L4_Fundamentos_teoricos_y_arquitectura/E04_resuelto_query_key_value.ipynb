## 1. Teoria breve

Query representa que busca el token actual. Key representa que ofrece cada token. Value es la informacion que se mezcla segun los pesos de attention.

## 2. Grafico guia

```mermaid
flowchart LR
    Q[Query: que busco] --> S[Score Q x K]
    K[Keys: que ofrece cada token] --> S
    S --> W[Pesos softmax]
    V[Values: informacion] --> O[Suma ponderada]
    W --> O
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
import math

query = [1.0, 0.2]
keys = {
    "lesion": [0.9, 0.1],
    "delantero": [1.0, 0.2],
    "anoto": [0.8, 0.5],
    "victoria": [0.7, 0.6],
}

def dot(a, b):
    return sum(x * y for x, y in zip(a, b))

scores = {token: dot(query, key) / math.sqrt(len(query)) for token, key in keys.items()}
total = sum(math.exp(score) for score in scores.values())
weights = {token: math.exp(score) / total for token, score in scores.items()}

for token, weight in sorted(weights.items(), key=lambda item: item[1], reverse=True):
    print(f"{token:<10} attention={weight:.3f}")
````

## 4. Interpretacion

El token objetivo combina Values de los tokens mas relevantes. La salida no es una palabra aislada, sino una representacion contextualizada.
