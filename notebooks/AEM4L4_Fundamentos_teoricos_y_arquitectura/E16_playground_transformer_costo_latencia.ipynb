## Para que sirve

Este notebook muestra que el costo de attention crece fuerte cuando aumentan los tokens. Permite pegar frases y ver relaciones aproximadas, latencia estimada y recomendacion de diseno.

**Cuando mostrarlo:** despues de `E06_anatomia_del_bloque_transformer.ipynb` y `E07_por_que_attention_cuesta_n_cuadrado.ipynb`.

**Idea docente:** mas contexto puede ayudar, pero no es gratis.


```python
# Editar en clase con frases cortas, parrafos o fragmentos de contrato.
entradas = [
    "Revisar saldo",
    "Necesito informacion sobre interchange fee capping y AML thresholding",
    "El contrato exige confidencialidad postcontractual, auditoria mensual, reporte de incidentes y preaviso de rescision",
]
```

```python
import re

def estimar_tokens(texto):
    partes = re.findall(r"[A-Za-z]+|\d+|[^\w\s]", texto)
    total = 0
    for p in partes:
        total += 1
        if len(p) > 10:
            total += 1
        if any(ch.isupper() for ch in p[1:]):
            total += 1
    return total

def perfil_transformer(texto):
    n = estimar_tokens(texto)
    relaciones = n * n
    latencia_ms = 80 + relaciones * 0.08
    if n <= 12:
        recomendacion = "ok para demo directa"
    elif n <= 35:
        recomendacion = "controlar prompt y salida"
    else:
        recomendacion = "considerar resumen, chunks o RAG"
    return n, relaciones, latencia_ms, recomendacion
```

```python
print("| entrada | tokens estimados | relaciones N^2 | latencia estimada | recomendacion |")
print("|---|---:|---:|---:|---|")
for entrada in entradas:
    n, relaciones, latencia, recomendacion = perfil_transformer(entrada)
    print(f"| {entrada} | {n} | {relaciones} | {latencia:.0f} ms | {recomendacion} |")
```

```python
# Curva visual N^2 para clase.
for n in [5, 10, 20, 40, 80, 120]:
    relaciones = n * n
    barra = "#" * min(60, relaciones // 200)
    print(f"{n:>3} tokens -> {relaciones:>6} relaciones | {barra}")
```
