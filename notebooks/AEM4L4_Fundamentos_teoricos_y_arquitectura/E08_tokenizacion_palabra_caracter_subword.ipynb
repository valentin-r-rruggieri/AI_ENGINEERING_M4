## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **Word-level** | Cada palabra completa intenta ser token. |
| **Character-level** | Cada caracter se procesa como token. |
| **Subword** | Divide palabras en piezas frecuentes. |
| **OOV** | Palabra fuera del vocabulario. |
| **UNK** | Token generico para algo desconocido. |
