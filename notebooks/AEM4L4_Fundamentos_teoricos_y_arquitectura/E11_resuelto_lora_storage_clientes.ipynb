## 1. Teoria breve

LoRA congela el modelo base e introduce matrices pequenas entrenables. La idea es que muchos cambios de dominio viven en un espacio de bajo rango, mas barato de almacenar y entrenar.

## 2. Grafico guia

```mermaid
    Runtime --> Response[Respuesta adaptada al dominio]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
model_gb = 14.0
adapter_gb = 0.33
clients = 50
full_ft = model_gb * clients
lora = model_gb + adapter_gb * clients

print(f"Full FT: {full_ft:.1f} GB")
print(f"LoRA   : {lora:.1f} GB")
print(f"Ahorro : {full_ft - lora:.1f} GB")
````

## 4. Interpretacion

LoRA es especialmente util en multi-tenant SaaS: una base compartida y adapters por cliente o dominio.

## 5. Errores comunes

- Creer que LoRA reemplaza evaluacion.
- Usar adapters con datos ruidosos sin control.
- No medir latencia de carga/cambio de adapter.
