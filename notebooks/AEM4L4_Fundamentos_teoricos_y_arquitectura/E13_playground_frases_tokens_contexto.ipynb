## Para que sirve

Este notebook es para abrirlo en vivo y cambiar la lista `frases`. Sirve para mostrar que una persona ve texto, pero el modelo procesa piezas. No usa un tokenizer real: usa una heuristica didactica para razonar antes de entrar en BPE, WordPiece y Transformer.

**Cuando mostrarlo:** despues de `E02_texto_tokens_y_contexto_desde_cero.ipynb`.

**Idea docente:** si una frase tiene terminos tecnicos, palabras largas o mezcla de idiomas, probablemente genere mas piezas y exija mas contexto.


```python
# Editar esta lista en clase.
frases = [
    "El banco aprobo el credito del cliente",
    "Me sente en el banco de la plaza",
    "Necesito informacion sobre interchange fee capping",
    "El paciente presenta hipercolesterolemia persistente",
    "El contrato exige confidencialidad postcontractual",
]
```

```python
import re

terminos_tecnicos = {
    "interchange", "capping", "hipercolesterolemia", "postcontractual",
    "confidencialidad", "thresholding", "settlement", "compliance",
}

def palabras(texto):
    return re.findall(r"[A-Za-z]+|\d+|[^\w\s]", texto)

def estimar_tokens(texto):
    total = 0
    detalle = []
    for p in palabras(texto):
        base = 1
        if len(p) > 10:
            base += 1
        if p.lower() in terminos_tecnicos:
            base += 1
        total += base
        detalle.append((p, base))
    return total, detalle

print("| # | frase | palabras | tokens estimados | alerta |")
print("|---|---|---:|---:|---|")
for i, frase in enumerate(frases, 1):
    toks, detalle = estimar_tokens(frase)
    cant_palabras = len(palabras(frase))
    alerta = "ok"
    if toks > cant_palabras + 2:
        alerta = "posible fragmentacion"
    if toks > 12:
        alerta = "revisar contexto/costo"
    print(f"| {i} | {frase} | {cant_palabras} | {toks} | {alerta} |")
```

```python
# Barras visuales para comparar longitud relativa.
for frase in frases:
    toks, _ = estimar_tokens(frase)
    barra = "#" * toks
    print(f"{toks:02d} tokens | {barra} | {frase}")
```
