# AEM4L4 E02 — RNN/LSTM vs Transformer

**Objetivo:** Comparar procesamiento secuencial contra procesamiento paralelo con attention.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecuta las celdas en orden.

## 1. Teoria breve

RNN y LSTM leen paso a paso. Eso dificulta recordar dependencias lejanas y limita paralelizacion. Transformer permite que cada posicion mire a muchas posiciones al mismo tiempo, lo que mejora relaciones largas y uso de hardware paralelo.

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
tokens = "A pesar de su lesion el delantero anoto el gol".split()
secuencial = list(enumerate(tokens, start=1))
pares_attention = [(a, b) for a in tokens for b in tokens]

print("Pasos secuenciales:", len(secuencial))
print("Relaciones posibles con attention:", len(pares_attention))
print("Ejemplo de relaciones:", pares_attention[:8])
````

## 4. Interpretacion

La ventaja no es que Transformer ignore el orden, sino que no depende de leer un token tras otro para conectar partes lejanas.
