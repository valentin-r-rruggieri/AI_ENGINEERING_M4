## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **RNN** | Red que procesa secuencias paso a paso. |
| **LSTM** | RNN con memoria mejorada para informacion relevante. |
| **Transformer** | Arquitectura que procesa posiciones en paralelo con attention. |
| **Dependencia larga** | Relacion entre partes alejadas de un texto. |
| **Paralelizacion** | Procesar muchas posiciones al mismo tiempo. |
