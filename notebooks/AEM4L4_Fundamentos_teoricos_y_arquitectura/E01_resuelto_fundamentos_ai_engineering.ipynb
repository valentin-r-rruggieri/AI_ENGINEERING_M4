## 1. Teoria breve

Un sistema con LLM no es solo un modelo: es una cadena de decisiones. El texto se transforma en tokens, los tokens se representan como embeddings, attention conecta contexto, el modelo se adapta a un dominio y la inferencia debe respetar costo y latencia.

## 2. Grafico guia

```mermaid
flowchart LR
    A[Texto del usuario] --> B[Tokenizacion]
    B --> C[Embeddings]
    C --> D[Bloques Transformer]
    D --> E[Respuesta]
    F[Dominio y datos] --> G[PEFT o LoRA]
    G --> D
    H[SLA y costo] --> I[Control de contexto]
    I --> B
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
decisiones = [
    ("arquitectura", "Transformer moderado"),
    ("tokenizacion", "subword robusta"),
    ("adaptacion", "LoRA si hay dominio especifico"),
    ("latencia", "limitar contexto y medir con casos reales"),
]

for decision, ejemplo in decisiones:
    print(f"{decision:<14} -> {ejemplo}")
````

## 4. Interpretacion

La clase completa gira alrededor de justificar decisiones. Elegir el modelo mas grande no alcanza si el tokenizer fragmenta demasiado, si el contexto se dispara o si adaptar el modelo completo es inviable.

## 5. Errores comunes

- Pensar que mas parametros siempre resuelven el problema.
- Ignorar tokenizacion y medir solo prompts cortos.
- No conectar decisiones tecnicas con costo, SLA y calidad.
