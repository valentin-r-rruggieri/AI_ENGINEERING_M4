## 1. Definiciones base

Antes de programar o discutir herramientas, fijamos vocabulario. El objetivo es que puedas explicar cada palabra tecnica con una frase simple.

| Concepto | Definicion simple |
|---|---|
| **Multi-head attention** | Varias atenciones en paralelo. |
| **FFN** | Red que refina cada posicion despues de attention. |
| **Residual** | Conexion que suma la entrada para no perder informacion. |
| **LayerNorm** | Normalizacion que estabiliza valores internos. |
| **Bloque** | Unidad repetida muchas veces dentro del Transformer. |
