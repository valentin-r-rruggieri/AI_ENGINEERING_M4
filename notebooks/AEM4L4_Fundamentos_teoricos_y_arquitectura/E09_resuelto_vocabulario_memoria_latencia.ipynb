## 1. Teoria breve

Un vocabulario grande puede reducir fragmentacion, pero aumenta matriz de embeddings. Un vocabulario chico ahorra memoria, pero puede producir mas tokens por texto y subir el costo de attention.

## 2. Grafico guia

```mermaid
flowchart TB
    VG[Vocabulario grande] --> VG1[Mas memoria en embeddings]
    VG --> VG2[Menos fragmentacion]
    VC[Vocabulario chico] --> VC1[Menos memoria]
    VC --> VC2[Mas tokens por palabra rara]
    VC2 --> L[Mas pares attention]
    VG2 --> L2[Menos pares attention]
```

## 3. Ejemplo guiado

La celda siguiente es intencionalmente pequena para poder explicarla en vivo antes de pasar a ejercicios combinados.

````python
configs = [
    {"name": "vocab_chico", "vocab": 30000, "tokens": 1200, "dim": 4096},
    {"name": "vocab_grande", "vocab": 100000, "tokens": 850, "dim": 4096},
]

for cfg in configs:
    embedding_params = cfg["vocab"] * cfg["dim"]
    attention_pairs = cfg["tokens"] ** 2
    print(cfg["name"])
    print("  embedding params:", f"{embedding_params:,}")
    print("  attention pairs :", f"{attention_pairs:,}")
````

## 4. Interpretacion

No hay respuesta universal: se decide segun dominio, hardware, latencia objetivo y textos reales.

## 5. Errores comunes

- Optimizar solo memoria y olvidar latencia.
- Optimizar solo tokens y olvidar embeddings.
- No medir con corpus del dominio.
