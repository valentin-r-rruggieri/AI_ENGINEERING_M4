# E04 - Resuelto: elegir transporte

## Que vas a practicar

Vas a decidir si conviene STDIO o HTTP segun el tipo de aplicacion.


## Tabla guia

| Caso | Transporte | Motivo |
|---|---|---|
| Antigravity local | STDIO | lanza un proceso local |
| VS Code workspace | STDIO | server del workspace |
| Host Python de clase | STDIO | proceso local controlado |
| SaaS multiusuario | HTTP | server remoto compartido |


## Diagrama textual

```text
Caso local con proceso propio -> STDIO
Caso remoto o compartido      -> HTTP
```


````python
def choose_transport(local, shared, remote):
    if local and not shared:
        return "STDIO"
    if remote or shared:
        return "HTTP"
    return "STDIO"

cases = [
    ("Antigravity", True, False, False),
    ("SaaS multiusuario", False, True, True),
]

for name, local, shared, remote in cases:
    print(name, "=>", choose_transport(local, shared, remote))
````

## Solucion explicada

Para nuestro MCP GitHub local usamos STDIO porque Antigravity puede lanzar `e01_github_mcp_server.py` como proceso.
