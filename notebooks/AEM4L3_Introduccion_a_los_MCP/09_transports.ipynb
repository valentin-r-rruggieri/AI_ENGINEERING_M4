# 09 - Transportes: STDIO vs HTTP

Objetivo: explicar por que para Antigravity y VS Code usamos STDIO local.


## Grafico guia

```mermaid
flowchart TD
    A["Cliente MCP"] --> B{"Transporte"}
    B --> C["STDIO local"]
    B --> D["HTTP remoto"]
    C --> E["Ideal para Antigravity/VS Code local"]
    D --> F["Ideal para servicios compartidos"]
```


````python
transport_table = [
    {"transport": "STDIO", "best_for": "local clients, desktop IDEs, teaching demos", "example": "VS Code starts python e01_github_mcp_server.py"},
    {"transport": "HTTP", "best_for": "remote service, shared infra", "example": "central MCP endpoint"},
]
transport_table
````

````python
def stdio_call(tool, args):
    return {"transport": "STDIO", "process": "local python", "tool": tool, "args": args}

def http_call(tool, args):
    return {"transport": "HTTP", "url": f"/mcp/tools/{tool}", "json": args}

stdio_call("github_create_repo", {"name": "aem4-demo"}), http_call("github_create_repo", {"name": "aem4-demo"})
````

## Donde se ve despues

Antigravity y VS Code levantan el mismo archivo Python por STDIO. El server lee stdin/stdout; el usuario no abre un puerto HTTP.
