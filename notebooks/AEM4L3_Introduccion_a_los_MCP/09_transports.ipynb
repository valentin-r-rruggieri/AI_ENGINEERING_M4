# 09 - Transportes: STDIO vs HTTP

## Que vas a aprender

Vas a elegir transporte segun el contexto. Para nuestra demo local usamos STDIO.


## Teoria de la lecture

El transporte define como se comunican Client y Server.

| Transporte | Conviene cuando | Ejemplo |
|---|---|---|
| STDIO | el server corre local como subproceso | Antigravity local, VS Code, host Python |
| HTTP | el server es remoto o compartido | SaaS multi-tenant |
| HTTP streaming | hay respuestas largas o eventos | servicios remotos con streaming |

STDIO no es inferior. Es ideal para integraciones locales.


## Diagrama textual

```text
STDIO local:
Antigravity -> lanza python e01_github_mcp_server.py -> stdin/stdout

HTTP remoto:
Host -> requests/http streaming -> MCP Server desplegado
```


````python
cases = {
    "Antigravity local": "STDIO",
    "VS Code workspace": "STDIO",
    "Host Python de clase": "STDIO",
    "SaaS compartido por muchos usuarios": "HTTP",
}

for case, transport in cases.items():
    print(f"{case}: {transport}")
````

## Paso que construimos del MCP GitHub

El server real termina con `mcp.run()`, que lo deja listo para comunicarse por STDIO.
