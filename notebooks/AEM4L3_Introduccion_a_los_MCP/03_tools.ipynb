# 03 - Tools: acciones invocables

## Que vas a aprender

Vas a distinguir una tool de lectura de una tool con side effect. Esto importa porque no todas requieren el mismo nivel de control.


## Teoria de la lecture

Una tool MCP es una accion invocable por el Host a traves del Client. Puede consultar datos o modificar un sistema.

| Tool | Modifica GitHub? | Riesgo | Control recomendado |
|---|---:|---|---|
| `github_get_repo` | no | bajo | validar owner/repo |
| `github_create_repo` | si | alto | repo privado, permisos, auditoria |
| `github_upsert_file` | si | alto | commit message, branch, auditoria |

Las tools con side effects no deberian existir sin contrato ni auditoria.


## Diagrama textual

```text
call_tool("github_create_repo", args)
  -> valida input
  -> llama create_repo(...)
  -> GitHub crea repo privado
  -> devuelve GitHubRepoResult
  -> escribe audit log
```


````python
tools = [
    {"name": "github_get_repo", "side_effect": False, "risk": "low"},
    {"name": "github_create_repo", "side_effect": True, "risk": "high"},
    {"name": "github_upsert_file", "side_effect": True, "risk": "high"},
]

for tool in tools:
    action = "requiere auditoria" if tool["side_effect"] else "puede ser consulta controlada"
    print(f"{tool['name']}: riesgo={tool['risk']} -> {action}")
````

## Paso que construimos del MCP GitHub

En el server real, estas tools aparecen con `@mcp.tool()`.

## Practica

Explica por que `github_create_repo` debe tener `private=True` por defecto. Despues agrega una regla para evitar tools peligrosas como `github_delete_repo`.
