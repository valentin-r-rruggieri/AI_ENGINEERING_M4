# 03 - Tools: funciones invocables por el modelo

Objetivo: distinguir tools puras de tools con side effect usando GitHub.


## Grafico guia

````python
def suggest_repo_name(project_name):
    normalized = project_name.lower().replace(" ", "-")
    return normalized[:40]

suggest_repo_name("AEM4 L3 MCP Demo")
````

La funcion anterior es pura: no toca sistemas externos. En cambio, crear un repo o subir un archivo tiene side effects reales.


````python
def plan_create_repo(name, description, private=True):
    return {
        "tool": "github_create_repo",
        "arguments": {"name": name, "description": description, "private": private},
        "side_effect": "creates_repository",
    }

plan_create_repo("aem4l3-mcp-demo", "Repositorio demo para clase")
````

````python
tools = [
    {
        "name": "github_create_repo",
        "description": "Creates a GitHub repository",
        "input_schema": {"name": "string", "description": "string", "private": "boolean"},
        "output_schema": {"full_name": "string", "html_url": "string"},
        "side_effect": True,
    },
    {
        "name": "github_upsert_file",
        "description": "Creates or updates a file with a commit",
        "input_schema": {"owner": "string", "repo": "string", "path": "string", "content": "string"},
        "output_schema": {"commit_sha": "string", "content_url": "string"},
        "side_effect": True,
    },
]
tools
````

## Donde se ve despues

En `e01_github_mcp_server.py`, estas tools aparecen como funciones decoradas con `@mcp.tool()`.
