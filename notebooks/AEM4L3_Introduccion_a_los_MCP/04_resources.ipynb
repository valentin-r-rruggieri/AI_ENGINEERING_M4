# 04 - Resources: contexto controlado

## Que vas a aprender

Vas a usar resources como datos de lectura que ayudan al Host sin ejecutar acciones reales.


## Teoria de la lecture

Un resource MCP informa. No deberia modificar sistemas externos. Tampoco deberia exponer secretos.

| Resource | Que informa | Que NO debe exponer |
|---|---|---|
| `github://config` | transporte, token configurado, audit path | `GITHUB_TOKEN` |
| `github://capabilities` | tools, resources, prompts y riesgos | secretos |
| `github://security-policy` | reglas operativas | credenciales |
| `github://audit/recent` | ultimos eventos | token o contenido sensible |
| `github://templates/readme-basic` | template reusable | datos privados |


## Diagrama textual

```text
read_resource("github://config")
  -> devuelve estado seguro
  -> no llama una tool
  -> no crea repos
  -> no muestra tokens
```


````python
resources = {
    "github://config": {"token_configured": True, "token_value": "NO_SE_EXPONE"},
    "github://capabilities": {"teaching_rule": "Tool ejecuta; resource informa; prompt guia."},
    "github://security-policy": {"destructive_tools_enabled": False},
}

for uri, payload in resources.items():
    print(uri, "=>", payload)
````

## Paso que construimos del MCP GitHub

Agregamos resources para que Antigravity pueda pedir contexto antes de ejecutar tools.

## Practica

Disena un resource `github://limits` que informe limites de uso, por ejemplo: maximo de repos por demo o ramas permitidas.
