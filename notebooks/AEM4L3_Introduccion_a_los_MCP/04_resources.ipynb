# 04 - Resources: contexto de lectura controlado

Objetivo: mostrar que no todo debe ser una tool. Algunos datos solo se leen.


## Grafico guia

```mermaid
flowchart LR
    A["Host"] --> B["Lee resource"]
```

````python
resources = {
    "github://config": {
        "server_name": "AEM4L3 GitHub MCP",
        "transport": "stdio",
        "token_configured": True,
        "default_private_repos": True,
        "dangerous_tools_exposed": [],
    },
    "github://repo-template": "README template for a new repository.",
}

def list_resources():
    return list(resources.keys())

def read_resource(uri):
    if uri not in resources:
        return {"error": "resource_not_found", "uri": uri}
    return {"uri": uri, "content": resources[uri]}

list_resources()
````

````python
read_resource("github://config")
````
