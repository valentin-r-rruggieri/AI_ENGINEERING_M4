## Que vas a practicar

Vas a decidir si un cambio al contrato de una tool es aditivo o breaking.


## Tabla guia

| Pregunta | Si la respuesta es si |
|---|---|
| Cambia el nombre de un campo existente? | breaking |
| Cambia el tipo de un campo existente? | breaking |
| Agrega un campo obligatorio? | breaking |
| Agrega un campo opcional? | aditivo |
| Agrega un campo al output? | aditivo |


## Diagrama textual

```text
Cambio de contrato
  -> aditivo: convive con clientes actuales
  -> breaking: requiere v2 o migracion
```


````python
changes = [
    ("agregar branch opcional", {"optional": True}),
    ("renombrar path a file_path", {"rename": True}),
    ("agregar content_url al output", {"output_only": True}),
]

for label, flags in changes:
    breaking = flags.get("rename") or flags.get("type_change") or flags.get("required")
    print(label, "=>", "breaking" if breaking else "aditivo")
````

## Solucion explicada

`branch` opcional y `content_url` en output son aditivos. Renombrar `path` rompe clientes existentes.
