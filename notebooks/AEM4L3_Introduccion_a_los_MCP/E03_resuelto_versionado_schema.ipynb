# AEM4L3 E03 — Tools, resources y prompts

Objetivo: entender las tres primitivas principales de MCP antes de ver el server real.


## Regla simple

- Tool: ejecuta una accion o computo. Puede tener side effects.
- Resource: aporta contexto de lectura. No deberia modificar sistemas.
- Prompt: plantilla reutilizable para guiar al LLM.


```python
capacidades = [
    {"nombre": "github_create_repo", "tipo": "tool", "riesgo": "alto: crea recurso real"},
    {"nombre": "github://config", "tipo": "resource", "riesgo": "bajo: solo lectura"},
    {"nombre": "repo_bootstrap_prompt", "tipo": "prompt", "riesgo": "medio: guia al modelo"},
]
capacidades
```

## Pregunta de clase

¿Por que `github_upsert_file` es una tool y no un resource? Porque produce un commit real. Esa diferencia define cuanta validacion y auditoria necesitamos.


```python
def clasificar(nombre, side_effect):
    return "tool" if side_effect else "resource o prompt"

clasificar("github_upsert_file", side_effect=True)
```
