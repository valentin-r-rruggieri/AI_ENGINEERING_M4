Objetivo: decidir si un cambio en una tool GitHub rompe clientes.


## Grafico guia

```mermaid
flowchart LR
    A["Schema v1"] --> B["Campo nuevo"]
    B --> C{"Rompe clientes?"}
```

````python
changes = [
    {"change": "agregar private=True", "type": "additive", "reason": "tiene default"},
    {"change": "renombrar name a repository_name", "type": "breaking", "reason": "rompe clientes viejos"},
    {"change": "agregar visibility sin default", "type": "breaking", "reason": "nuevo campo obligatorio"},
    {"change": "agregar html_url al output", "type": "additive", "reason": "clientes viejos lo pueden ignorar"},
]
changes
````

````python
[c for c in changes if c["type"] == "breaking"]
````
