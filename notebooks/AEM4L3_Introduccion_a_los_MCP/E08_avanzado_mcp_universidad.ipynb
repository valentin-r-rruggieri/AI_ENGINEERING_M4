| Capa | Pregunta | Ejemplo GitHub |

| Resources | Que contexto se lee? | config, policy, audit |
| Prompts | Que instrucciones se reutilizan? | bootstrap, safety review |
| Auditoria | Que se registra? | tool, repo, timestamp |
