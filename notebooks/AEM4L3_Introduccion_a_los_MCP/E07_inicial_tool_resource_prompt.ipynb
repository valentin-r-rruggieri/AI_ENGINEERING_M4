# AEM4L3 E07 — Seguridad: scopes, tenant isolation y confused deputy

Objetivo: mostrar que MCP Server es frontera de seguridad, no solo catalogo de tools.


## Conceptos

- Scope: permiso minimo para ejecutar una tool.
- Tenant isolation: un cliente no puede leer/escribir datos de otro.
- Confused deputy: el server usa privilegios propios para ejecutar algo que el usuario no podia hacer.
- Audit log: evidencia de quien pidio que y que decidio el server.


```python
contexto_confiable = {
    "tenant_id": "tenant_demo",
    "actor_id": "instructor",
    "scopes": ["tickets:read", "tickets:write", "policies:read"],
}
body_no_confiable = {"tenant_id": "otro_tenant", "customer_id": "cus_123"}
contexto_confiable, body_no_confiable
```

## Lectura docente

El tenant confiable debe venir del token/contexto, no del body que manda el cliente. Si una tool acepta `tenant_id` libre y lo cree, aparece riesgo de confused deputy.


```python
def decision_tenant(contexto, body):
    return {
        "usar": contexto["tenant_id"],
        "rechazar": body["tenant_id"],
        "motivo": "tenant viene de contexto confiable",
    }

decision_tenant(contexto_confiable, body_no_confiable)
```
