# 02 - MCP Server minimo como catalogo de capacidades

Objetivo: entender que un server MCP no es solo una funcion: publica un catalogo descubrible.


## Grafico guia

```mermaid
flowchart TD
    A["Definir server"] --> B["Registrar tool"]
    B --> C["Registrar resource"]
    C --> D["Registrar prompt"]
    D --> E["Ejecutar por STDIO"]
```


````python
server_catalog = {
    "name": "AEM4L3 GitHub MCP",
    "transport": "stdio",
    "tools": [],
    "resources": [],
    "prompts": [],
}
server_catalog
````

````python
server_catalog["tools"].append({
    "name": "github_create_repo",
    "description": "Create a private GitHub repository",
    "version": "1.0.0",
    "side_effect": True,
})
server_catalog["resources"].append({
    "uri": "github://config",
    "description": "Safe server configuration without exposing token",
})
server_catalog["prompts"].append({
    "name": "repo_bootstrap_prompt",
    "version": "1.0.0",
})
server_catalog
````

## Lectura docente

El server le dice al host que sabe hacer. El host no tiene que conocer GitHub internamente: descubre capacidades y llama contratos.
