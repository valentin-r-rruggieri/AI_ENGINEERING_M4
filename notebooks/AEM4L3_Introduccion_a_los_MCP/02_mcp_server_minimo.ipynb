# 02 - MCP Server minimo: catalogo de capacidades

## Que vas a aprender

Vas a entender que un MCP Server no es una funcion suelta. Es un catalogo descubrible que puede publicar tools, resources y prompts.


## Teoria de la lecture

Un cliente MCP primero descubre que capacidades existen. Esto evita que el Host tenga que hardcodear todos los detalles.

| Capacidad | Pregunta que responde |
|---|---|
| Tool | Que acciones puedo ejecutar? |
| Resource | Que contexto puedo leer? |
| Prompt | Que instrucciones reutilizables existen? |

En el server real usamos `FastMCP`, pero antes vamos a simular el catalogo con Python puro.


## Diagrama textual

```text
MCP Server GitHub
  tools:
    - github_create_repo
    - github_upsert_file
    - github_get_repo
  resources:
    - github://config
    - github://capabilities
  prompts:
    - repo_bootstrap_prompt
```

Ese catalogo es lo que despues ve Antigravity.


````python
server_catalog = {
    "name": "AEM4L3 GitHub MCP",
    "transport": "stdio",
    "tools": [],
    "resources": [],
    "prompts": [],
}

server_catalog["tools"].append({"name": "github_create_repo", "side_effect": True})
server_catalog["resources"].append({"uri": "github://config", "side_effect": False})
server_catalog["prompts"].append({"name": "repo_bootstrap_prompt", "side_effect": False})

print(server_catalog)
````

## Interpretacion

El server no ejecuto GitHub todavia. Solo publico que sabe hacer. Ese discovery es una diferencia central contra un wrapper ad hoc.

## Practica

Agrega al catalogo una tool llamada `github_upsert_file` y marcala como `side_effect=True`. Despues agrega un resource llamado `github://security-policy`.
