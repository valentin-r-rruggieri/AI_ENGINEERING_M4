# 05 - Prompts reutilizables

Objetivo: mostrar que un MCP server tambien puede publicar prompts estandarizados.


## Grafico guia

flowchart LR

````python
prompts = {
    "repo_bootstrap_prompt": {
        "version": "1.0.0",
        "template": """Actua como asistente de ingenieria.
Proyecto: {project_name}
Objetivo: {goal}
Primero crea un repo privado. Despues agrega un README.md con objetivo, estructura inicial y proximos pasos.
""",
    }
}

def list_prompts():
    return list(prompts.keys())

def render_prompt(name, **kwargs):
    return prompts[name]["template"].format(**kwargs)

list_prompts()
````

````python
print(render_prompt(
    "repo_bootstrap_prompt",
    project_name="aem4l3-mcp-demo",
    goal="mostrar un MCP propio conectado a GitHub",
))
````

## Donde se ve despues

En `e01_github_mcp_server.py`, `repo_bootstrap_prompt` esta decorado con `@mcp.prompt()`.
