# 05 - Prompts: instrucciones reutilizables

## Que vas a aprender

Vas a ver que un prompt MCP no ejecuta nada. Sirve para estandarizar como el LLM piensa o redacta una accion.


## Teoria de la lecture

Los prompts ayudan a que varios Hosts usen una misma instruccion de forma consistente.

| Prompt | Para que sirve | Ejecuta GitHub? |
|---|---|---:|
| `repo_bootstrap_prompt` | guiar repo inicial | no |
| `repo_readme_prompt` | redactar README | no |
| `safe_github_action_prompt` | revisar riesgos | no |
| `repo_review_prompt` | guiar revision | no |

La frase clave: **tool ejecuta; resource informa; prompt guia**.


## Diagrama textual

```text
get_prompt("safe_github_action_prompt", action_summary)
  -> devuelve una instruccion
  -> el LLM razona riesgos
  -> despues, si corresponde, se llama una tool
```


````python
def repo_readme_prompt(project_name, goal, audience="alumnos"):
    return (
        "Genera un README.md claro. "
        f"Proyecto: {project_name}. Objetivo: {goal}. Audiencia: {audience}. "
        "Inclui objetivo, estructura, como probarlo y proximos pasos."
    )

print(repo_readme_prompt("aem4l3-mcp-demo", "mostrar MCP con GitHub"))
````

## Paso que construimos del MCP GitHub

Publicamos prompts para que Antigravity pueda pedir instrucciones reutilizables antes de llamar tools.
