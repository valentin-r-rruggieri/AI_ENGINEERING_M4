# E01 - Resuelto: clasificar primitivas MCP

## Que vas a practicar

Vas a clasificar capacidades como tool, resource o prompt usando el MCP GitHub real como referencia.


## Recordatorio rapido

| Primitiva | Pregunta | Ejemplo |
|---|---|---|
| Tool | Ejecuta una accion? | `github_create_repo` |
| Resource | Informa contexto? | `github://config` |
| Prompt | Guia al modelo? | `repo_bootstrap_prompt` |


## Diagrama textual

```text
Capacidad MCP
  -> si ejecuta accion: Tool
  -> si entrega contexto: Resource
  -> si devuelve instruccion reusable: Prompt
```


````python
items = {
    "github_create_repo": "tool",
    "github_upsert_file": "tool",
    "github_get_repo": "tool",
    "github://config": "resource",
    "github://capabilities": "resource",
    "repo_bootstrap_prompt": "prompt",
    "safe_github_action_prompt": "prompt",
}

for name, primitive in items.items():
    print(f"{name}: {primitive}")
````

## Solucion explicada

- `github_create_repo` y `github_upsert_file` son tools porque ejecutan acciones reales.
- `github://config` y `github://capabilities` son resources porque informan contexto.
- `repo_bootstrap_prompt` y `safe_github_action_prompt` son prompts porque devuelven instrucciones reutilizables.
