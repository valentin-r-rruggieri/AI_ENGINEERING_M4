# AEM4L3 E01 — Problema ad hoc y por que MCP

Objetivo: ver que el problema inicial no es el modelo, sino tener muchas integraciones distintas para la misma capacidad. Este notebook prepara el terreno para `e01_tool_calling_ad_hoc_vs_mcp.py`.


## Concepto

Una integracion ad hoc conecta una app con una herramienta de forma puntual. Funciona en demo, pero cuando aparecen varios clientes, wrappers y permisos, la arquitectura se fragmenta. MCP propone publicar capacidades en un servidor comun: tools para acciones, resources para contexto y prompts para instrucciones reutilizables.


```python
wrappers = {
    "slack": {"function": "create_ticket", "fields": ["order_id", "reason"]},
    "web": {"function": "createTicket", "fields": ["orderNumber", "notes"]},
    "desktop": {"function": "new_case", "fields": ["id", "description"]},
}
wrappers
```

## Lectura docente

Los tres wrappers quieren crear un ticket, pero cada uno usa nombres distintos. Esta diferencia parece menor hasta que una app valida permisos, otra no, una guarda logs y otra no. Ahi MCP se vuelve una decision de arquitectura, no una moda.


```python
contrato_mcp = {
    "tool": "create_ticket",
    "input_schema": {"order_id": "string", "reason": "string"},
    "output_schema": {"ticket_id": "string", "status": "string"},
    "required_scope": "tickets:write",
}
contrato_mcp
```
