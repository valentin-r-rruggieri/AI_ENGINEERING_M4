## Definiciones

- **Host:** la aplicacion donde vive la conversacion: Antigravity, VS Code o nuestro script OpenAI.
- **Client:** la pieza que habla el protocolo MCP con un server.
- **Server:** el proceso que publica tools, resources y prompts.


````python
host = {
    "name": "VS Code / Antigravity / OpenAI host",
    "responsibility": "entiende el pedido del usuario y decide que capacidad usar",
}
client = {
    "name": "MCP Client",
    "responsibility": "inicializa sesion, lista tools y llama al server",
}
server = {
    "name": "AEM4L3 GitHub MCP",
    "responsibility": "ejecuta GitHub con contratos y auditoria",
}
host, client, server
````

````python
flow = [
    "USER pide crear un repo privado",
    "HOST interpreta el pedido",
    "CLIENT lista tools del server MCP",
    "SERVER publica github_create_repo y github_upsert_file",
    "HOST elige una tool",
    "CLIENT llama la tool por STDIO",
    "SERVER ejecuta GitHub API y audita",
]
flow
````
