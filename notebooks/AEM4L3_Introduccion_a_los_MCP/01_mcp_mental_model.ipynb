# 01 - Modelo mental MCP: Host, Client y Server

## Que vas a aprender

Vas a separar las tres piezas principales de MCP: Host, Client y Server. Esta separacion es la base para entender Antigravity, VS Code o nuestro host Python con OpenAI.


## Teoria de la lecture

MCP no dice que el LLM pueda llamar cualquier API directamente. MCP separa responsabilidades.

| Pieza | Responsabilidad | En nuestro ejemplo |
|---|---|---|
| Host | Entiende al usuario y decide que hacer | Antigravity o `e02_openai_host_usa_mcp_github.py` |
| MCP Client | Habla el protocolo MCP | `ClientSession` por STDIO |
| MCP Server | Publica tools, resources y prompts | `e01_github_mcp_server.py` |
| Sistema externo | API real que se quiere usar | GitHub REST API |

La clave: el Host consume capacidades; el Server las gobierna.


## Diagrama textual

```text
Usuario
  -> Host con LLM
      -> MCP Client
          -> MCP Server GitHub
              -> GitHub API
```

El usuario no toca GitHub directamente. El LLM tampoco deberia tener acceso libre. El server MCP marca el borde seguro.


## Paso que construimos del MCP GitHub

Identificamos que nuestro server se llamara `AEM4L3 GitHub MCP` y que publicara:

| Tipo | Ejemplo |
|---|---|
| Tool | `github_create_repo` |
| Resource | `github://config` |
| Prompt | `repo_bootstrap_prompt` |


````python
host = {
    "name": "Antigravity / VS Code / Host Python con OpenAI",
    "job": "entiende el pedido y decide que capacidad usar",
}
client = {
    "name": "MCP Client",
    "job": "inicializa sesion, lista capacidades y llama al server",
}
server = {
    "name": "AEM4L3 GitHub MCP",
    "job": "publica capacidades gobernadas para GitHub",
}
external_api = {"name": "GitHub REST API", "job": "crea repos, lee metadata y escribe archivos"}

for piece in [host, client, server, external_api]:
    print(f"{piece['name']}: {piece['job']}")
````
