## Idea clave

Un contrato MCP no es decoracion. Define input, output, version, permisos y errores esperados. Si el contrato es ambiguo, el cliente no sabe como invocar ni como recuperarse.


```python
from pydantic import BaseModel, Field, ValidationError
from typing import Literal

class CreateRepoArgs(BaseModel):
    name: str = Field(..., min_length=3)
    description: str
    private: bool = True

CreateRepoArgs(name="aem4l3-demo", description="Repo de clase")
```

```python
try:
    CreateRepoArgs(name="x", description="Nombre demasiado corto")
except ValidationError as exc:
    print(exc)
```
