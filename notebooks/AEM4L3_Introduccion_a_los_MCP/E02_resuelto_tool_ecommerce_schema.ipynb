# E02 - Resuelto: schema de una tool

## Que vas a practicar

Vas a construir un schema simple y compararlo con la tool `github_upsert_file`.


## Teoria breve

Un schema evita que el Host mande cualquier payload. Sirve para validar antes de ejecutar.

| Campo | Tipo | Obligatorio |
|---|---|---:|
| `owner` | string | si |
| `repo` | string | si |
| `path` | string | si |
| `content` | string | si |
| `commit_message` | string | si |
| `branch` | string | no |


## Diagrama textual

```text
Payload del Host
  -> schema de github_upsert_file
      -> owner, repo, path, content, commit_message
      -> branch opcional
  -> si valida: puede llamarse la tool
```


````python
schema = {
    "owner": {"type": "str", "required": True},
    "repo": {"type": "str", "required": True},
    "path": {"type": "str", "required": True},
    "content": {"type": "str", "required": True},
    "commit_message": {"type": "str", "required": True, "min_length": 5},
    "branch": {"type": "str", "required": False, "default": "main"},
}

payload = {
    "owner": "valen-ruggieri",
    "repo": "aem4l3-mcp-demo",
    "path": "README.md",
    "content": "# Demo",
    "commit_message": "Add README",
}

missing = [field for field, spec in schema.items() if spec["required"] and field not in payload]
print("missing:", missing)
print("payload listo para tool" if not missing else "faltan campos")
````

## Solucion explicada

El payload es valido porque tiene todos los campos obligatorios. `branch` puede omitirse porque tiene default `main`.
