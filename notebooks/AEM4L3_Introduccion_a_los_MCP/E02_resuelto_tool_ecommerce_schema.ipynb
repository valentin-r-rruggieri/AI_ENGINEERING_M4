Objetivo: separar responsabilidades. Este notebook prepara el terreno para `e04_github_mcp_server.py` y `e05_openai_host_usa_mcp_github.py`.


## Definiciones

- MCP Host: aplicacion donde vive la experiencia con LLM, por ejemplo Antigravity.
- MCP Client: componente que habla protocolo MCP con el server.
- MCP Server: proceso que publica tools, resources y prompts, valida entradas y ejecuta acciones.


```python
flujo = [
    "USER pide crear repo",
    "HOST usa OpenAI para decidir accion",
    "CLIENT llama tool MCP por STDIO",
    "SERVER valida y ejecuta GitHub",
    "SERVER devuelve structured output",
]
for paso in flujo:
    print(paso)
```

## Lectura docente

El LLM no deberia tener el token de GitHub ni conocer todos los detalles HTTP. El server MCP guarda esa frontera. El host conversa; el server ejecuta con reglas.
