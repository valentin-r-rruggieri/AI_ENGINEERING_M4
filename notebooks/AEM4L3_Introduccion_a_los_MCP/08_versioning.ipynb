# 08 - Versionado de tools

Objetivo: diferenciar cambios aditivos de breaking changes.


## Grafico guia

```mermaid
flowchart LR
    A["Tool v1"] --> B["Clientes existentes"]
    A --> C["Nueva necesidad"]
    C --> D["Tool v2"]
    D --> E["Compatibilidad y migracion"]
```


````python
repo_tool_v1 = {
    "name": "github_create_repo",
    "version": "1.0.0",
    "input": {"name": "string", "description": "string"},
}

repo_tool_v1_1 = {
    "name": "github_create_repo",
    "version": "1.1.0",
    "input": {"name": "string", "description": "string", "private": "boolean default true"},
}

repo_tool_v2 = {
    "name": "github_create_repository",
    "version": "2.0.0",
    "input": {"repository_name": "string", "visibility": "private|public"},
}

repo_tool_v1, repo_tool_v1_1, repo_tool_v2
````

````python
def classify_change(old_input, new_input):
    missing_old_fields = set(old_input) - set(new_input)
    if missing_old_fields:
        return "breaking"
    return "additive"

classify_change(repo_tool_v1["input"], repo_tool_v1_1["input"]), classify_change(repo_tool_v1["input"], repo_tool_v2["input"])
````
