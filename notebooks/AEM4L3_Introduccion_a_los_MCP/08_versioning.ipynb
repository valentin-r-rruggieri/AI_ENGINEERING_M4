# 08 - Versionado: cambios aditivos y breaking changes

## Que vas a aprender

Vas a clasificar cambios de contrato para no romper Hosts que ya usan el MCP.


## Teoria de la lecture

Cuando una tool ya esta publicada, cambiar su contrato puede romper clientes.

| Cambio | Tipo | Por que |
|---|---|---|
| Agregar campo opcional | aditivo | clientes viejos siguen funcionando |
| Agregar campo nuevo al output | aditivo | el cliente puede ignorarlo |
| Renombrar un input | breaking | el cliente viejo manda el nombre anterior |
| Cambiar tipo de un campo | breaking | rompe validacion |
| Agregar obligatorio | breaking | el cliente viejo no lo envia |


## Diagrama textual

```text
Contrato v1 -> clientes actuales
     |
     +-- cambio aditivo -> puede convivir
     |
     +-- breaking change -> crear v2 o plan de migracion
```


````python
changes = [
    {"change": "agregar branch opcional", "required": False, "renames": False, "type_changed": False},
    {"change": "renombrar path a file_path", "required": False, "renames": True, "type_changed": False},
    {"change": "agregar reviewer obligatorio", "required": True, "renames": False, "type_changed": False},
]

for item in changes:
    breaking = item["required"] or item["renames"] or item["type_changed"]
    print(item["change"], "=>", "BREAKING" if breaking else "ADITIVO")
````

## Paso que construimos del MCP GitHub

Nuestro `github_upsert_file` tiene `branch="main"` como opcional. Eso es aditivo y permite reutilizar la tool sin forzar a todos los clientes a mandar branch.
