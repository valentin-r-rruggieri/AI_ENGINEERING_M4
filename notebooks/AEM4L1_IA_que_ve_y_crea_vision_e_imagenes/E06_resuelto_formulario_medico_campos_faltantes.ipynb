````python
from pydantic import BaseModel, ValidationError
from datetime import date
from typing import Optional

print("=" * 60)
print("Caso de uso: Formulario médico con campos incompletos")
print("=" * 60)

class PatientForm(BaseModel):
    """
    Modelo para formulario médico.
    
    - patient_name: OBLIGATORIO
    - document_id: OBLIGATORIO
    - birth_date: OBLIGATORIO
    - insurance: OPCIONAL (puede ser None si el paciente no tiene)
    - signature_present: OBLIGATORIO booleano (¿tiene firma?)
    """
    patient_name: str
    document_id: str
    birth_date: date
    insurance: Optional[str] = None
    signature_present: bool

print("\n✅ Caso 1: Formulario COMPLETO con seguro:")
llm_output_1 = {
    "patient_name": "Lucía Fernández",
    "document_id": "38200123",
    "birth_date": "1998-04-20",
    "insurance": "MEDIFE",
    "signature_present": True
}

form1 = PatientForm(**llm_output_1)
print(f"   ✅ {form1}")

print("\n✅ Caso 2: Formulario INCOMPLETO sin seguro:")
llm_output_2 = {
    "patient_name": "Juan Pérez",
    "document_id": "25500456",
    "birth_date": "1980-06-15",
    "insurance": None,  # NULL/None - campo opcional vacío
    "signature_present": False  # Sin firma
}

form2 = PatientForm(**llm_output_2)
print(f"   ✅ {form2}")
print(f"   - Seguro: {form2.insurance}")
print(f"   - Firma presente: {form2.signature_present}")

print("\n❌ Caso 3: Intento sin campo OBLIGATORIO:")
try:
    llm_output_3 = {
        # Falta patient_name
        "document_id": "30123789",
        "birth_date": "1995-03-10",
        "signature_present": True
    }
    form3 = PatientForm(**llm_output_3)
except ValidationError as e:
    print(f"   ❌ Error: Campo obligatorio faltante")
    print(f"   {e.errors()[0]['msg']}")

print("\n💡 Lecciones:")
print("   ✅ Optional[str] = None: permite nulos pero valida si NO es nulo")
print("   ✅ bool: campos booleanos para estados Yes/No, True/False")
print("   ✅ date: conversión automática de strings YYYY-MM-DD")
````
