## Transformación: Texto → JSON

El objetivo es ir de:
- **Texto libre:** Difícil de procesar, desorganizado ⚠️
- **JSON estructurado:** Campos claros, fácil de procesar ✅

En un proyecto real, usarías un LLM (ChatGPT) con instrucción JSON_MODE o similar para generar automáticamente el JSON.

```python
import json

print("=" * 60)
print("PASO 2: Convertir texto a JSON estructurado")
print("=" * 60)

print("\n💡 En un proyecto real, usarías:")
print("   client.chat.completions.create(")
print("       model='gpt-4o',")
print("       messages=[{")
print("           'role': 'user',")
print("           'content': 'Convierte este texto a JSON: ...'")
print("       }],")
print("       response_format={'type': 'json_object'}")
print("   )")

print("\n" + "-" * 60)
print("PARA ESTE EJERCICIO: Simulamos el JSON resultante")
print("-" * 60)

# JSON estructurado que resultaría de procesar el texto
receipt_json = {
    "store_name": "Supermercado Norte",
    "purchase_date": "2026-06-12",
    "total_amount": 8500,
    "payment_method": "tarjeta de crédito"
}

print("\n📦 JSON estructurado:")
print(json.dumps(receipt_json, indent=2, ensure_ascii=False))

print("\n✅ Ventajas de esta estructura:")
print("   - Campos claramente definidos")
print("   - Tipos consistentes")
print("   - Fácil de procesar en el backend")
print("   - Listo para VALIDAR con Pydantic (próximo paso)")

print("\n" + "-" * 60)
print("PRÓXIMO PASO: E05 - Validar este JSON con Pydantic")
print("-" * 60)
```
