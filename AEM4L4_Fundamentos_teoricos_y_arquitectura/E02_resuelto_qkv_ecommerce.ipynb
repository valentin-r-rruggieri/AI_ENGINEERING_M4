**Objetivo:** Explicar Query, Key y Value con analogía.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Query es lo que busco; Key es etiqueta; Value es contenido útil.

```python
query = "zapatillas negras para correr"
keys = ["zapatillas", "running", "negras", "deporte"]
values = {"precio": 65000, "marca": "RunFast", "stock": 12}

print("Query:", query)
print("Keys relevantes:", keys)
print("Values recuperados:", values)
```
