# AEM4L5 E04 — Async para llamadas I/O

**Objetivo:** Comparar secuencial vs async.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Async reduce espera cuando el cuello de botella es I/O.

```python
import time, asyncio

def sync_call_api(i):
    time.sleep(1)
    return f"result {i}"

start = time.time()
results = [sync_call_api(i) for i in range(5)]
print("Sync:", results, "tiempo:", round(time.time() - start, 2))

async def async_call_api(i):
    await asyncio.sleep(1)
    return f"result {i}"

async def main():
    start = time.time()
    results = await asyncio.gather(*[async_call_api(i) for i in range(5)])
    print("Async:", results, "tiempo:", round(time.time() - start, 2))

await main()
```
