# AEM4L5 E02 — Elegir despliegue

**Objetivo:** Decidir serverless, persistente o híbrido.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

La decisión depende de tráfico, costo, cold starts y latencia.

```python
scenarios = {
    "10 requests por día": ("Serverless", "paga por uso"),
    "API 24/7": ("Servidor persistente", "baja latencia estable"),
    "Tráfico impredecible": ("Serverless/híbrido", "escala flexible"),
    "Chatbot tiempo real": ("Servidor persistente", "evita cold starts")
}

for s, decision in scenarios.items():
    print(s, "→", decision)
```
