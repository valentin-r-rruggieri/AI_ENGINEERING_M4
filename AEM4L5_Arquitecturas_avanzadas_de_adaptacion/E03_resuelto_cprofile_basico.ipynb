# AEM4L5 E03 — cProfile básico

**Objetivo:** Perfilar antes de optimizar.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Se compara limpieza lenta vs regex.

```python
import cProfile, pstats, io, re

texts = ["Hola!!! Esto es una prueba 123"] * 5000

def clean_text_slow(text):
    result = ""
    for c in text:
        if c.isalnum() or c.isspace():
            result += c.lower()
    return result

def clean_text_fast(text):
    cleaned = re.sub(r"[^A-Za-z0-9\s]+", "", text)
    return cleaned.lower()

def process_batch(func):
    return [func(t) for t in texts]

for func in [clean_text_slow, clean_text_fast]:
    profile = cProfile.Profile()
    profile.enable()
    process_batch(func)
    profile.disable()
    s = io.StringIO()
    pstats.Stats(profile, stream=s).sort_stats("cumulative").print_stats(5)
    print("\n---", func.__name__, "---")
    print(s.getvalue())
```
