**Objetivo:** Combinar WER, resumen y confiabilidad.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Si WER supera el umbral, el resumen no debería considerarse confiable.

```python
def simple_wer(reference, hypothesis):
    ref = reference.lower().split()
    hyp = hypothesis.lower().split()
    errors = sum(1 for a, b in zip(ref, hyp) if a != b) + abs(len(ref) - len(hyp))
    return errors / len(ref)

cases = [
    {
        "reference": "el cliente solicita cancelar el pedido por demora",
        "hypothesis": "el cliente solicita cancelar el pedido por demora",
        "summary": "Cliente solicita cancelar pedido por demora."
    },
    {
        "reference": "el paciente debe tomar la medicación cada ocho horas",
        "hypothesis": "el paciente debe tomar la medicación cada dos horas",
        "summary": "Paciente debe tomar medicación cada dos horas."
    }
]

for c in cases:
    wer = simple_wer(c["reference"], c["hypothesis"])
    print({"wer": round(wer, 2), "summary": c["summary"], "is_reliable": wer < 0.20})
```
