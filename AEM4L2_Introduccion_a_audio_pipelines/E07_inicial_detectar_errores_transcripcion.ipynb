# AEM4L2 E07 — Inicial | Detectar errores

**Objetivo:** Clasificar match, substitution, deletion e insertion.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Ejercicio básico sin código complejo.

```python
pairs = [
    ("pedido", "perdido"),
    ("mañana", "mañana"),
    ("urgente", None),
    (None, "eh")
]

for ref, hyp in pairs:
    if ref == hyp:
        label = "match"
    elif ref is None:
        label = "insertion"
    elif hyp is None:
        label = "deletion"
    else:
        label = "substitution"
    print(ref, hyp, "→", label)
```
