# AEM4L2 E02 — Calcular WER manualmente

**Objetivo:** Calcular substitutions, deletions, insertions y WER.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

El WER mide proporción de errores frente a la referencia.

```python
reference = "Necesito cancelar mi turno médico de mañana"
hypothesis = "Necesito cambiar mi turno medico mañana"

S = 2  # cancelar→cambiar, médico→medico si no normalizamos acentos
D = 1  # de eliminado
I = 0
N = len(reference.split())
WER = (S + D + I) / N

print({"S": S, "D": D, "I": I, "N": N, "WER": WER})
```
