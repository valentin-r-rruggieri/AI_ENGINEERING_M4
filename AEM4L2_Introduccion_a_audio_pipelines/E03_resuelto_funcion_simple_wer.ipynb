# AEM4L2 E03 — Función simple de WER

**Objetivo:** Implementar una métrica básica para comparar transcripciones.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Versión simple para clase; luego puede mejorarse con distancia de edición.

```python
def simple_wer(reference: str, hypothesis: str) -> float:
    ref_words = reference.lower().split()
    hyp_words = hypothesis.lower().split()
    errors = sum(1 for a, b in zip(ref_words, hyp_words) if a != b)
    errors += abs(len(ref_words) - len(hyp_words))
    return errors / len(ref_words)

print(simple_wer("hola quiero cancelar mi pedido", "hola quiero cambiar mi pedido"))
```
