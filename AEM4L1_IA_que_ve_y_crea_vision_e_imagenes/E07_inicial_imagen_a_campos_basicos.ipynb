Solo dos campos: full_name y document_number.

```python
from pydantic import BaseModel

class BasicDocument(BaseModel):
    full_name: str
    document_number: str

output = {
    "full_name": "Juan Pérez",
    "document_number": "40111222"
}

print(BasicDocument(**output))
```
