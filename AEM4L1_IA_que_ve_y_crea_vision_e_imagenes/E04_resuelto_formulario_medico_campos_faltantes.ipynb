Este caso muestra `null` y campos booleanos visuales, como presencia de firma.

```python
from pydantic import BaseModel
from datetime import date
from typing import Optional

class PatientForm(BaseModel):
    patient_name: str
    document_id: str
    birth_date: date
    insurance: Optional[str] = None
    signature_present: bool

llm_output = {
    "patient_name": "Lucía Fernández",
    "document_id": "38200123",
    "birth_date": "1998-04-20",
    "insurance": None,
    "signature_present": True
}

print(PatientForm(**llm_output))
```
