```python
import pydantic
from pydantic import BaseModel, ValidationError
from datetime import date
from typing import Optional

class ReceiptData(BaseModel):
    store_name: str
    purchase_date: date
    total_amount: float
    payment_method: Optional[str] = None

valid_output = {
    "store_name": "Supermercado Norte",
    "purchase_date": "2026-06-12",
    "total_amount": 8500,
    "payment_method": "tarjeta"
}

print(ReceiptData(**valid_output))

invalid_output = {
    "store_name": "Supermercado Norte",
    "purchase_date": "12/06/2026",
    "total_amount": "ocho mil quinientos"
}

try:
    ReceiptData(**invalid_output)
except ValidationError as e:
    print("\nErrores de validación:")
    print(e)
```
