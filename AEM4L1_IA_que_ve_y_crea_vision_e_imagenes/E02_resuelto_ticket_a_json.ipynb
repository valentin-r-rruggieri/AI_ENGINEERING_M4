# AEM4L1 E02 — Ticket → JSON estructurado

**Objetivo:** Transformar texto extraído en JSON útil para backend.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

El objetivo es que el output deje de ser texto libre y pase a tener campos estables.

```python
import json

receipt_json = {
    "store_name": "Supermercado Norte",
    "purchase_date": "2026-06-12",
    "total_amount": 8500,
    "payment_method": "tarjeta"
}

print(json.dumps(receipt_json, indent=2, ensure_ascii=False))
```
