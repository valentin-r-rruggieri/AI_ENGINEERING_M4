# PIM4 E04 — Pydantic + error handling

**Objetivo:** Validar ContractChangeOutput.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Manejar outputs inválidos antes de producción.

```python
from pydantic import BaseModel, ValidationError, Field
from typing import List

class ContractChangeOutput(BaseModel):
    sections_changed: List[str]
    topics_touched: List[str]
    summary_of_the_change: str = Field(..., min_length=10)

valid = {
    "sections_changed": ["payment_terms"],
    "topics_touched": ["monto mensual"],
    "summary_of_the_change": "El monto mensual aumentó de $1000 a $1200."
}
print(ContractChangeOutput(**valid))

invalid = {
    "sections_changed": "payment_terms",
    "summary_of_the_change": ""
}

try:
    ContractChangeOutput(**invalid)
except ValidationError as e:
    print("\nErrores:")
    print(e)
```
