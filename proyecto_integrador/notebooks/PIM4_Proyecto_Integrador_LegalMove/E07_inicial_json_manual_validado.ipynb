**Objetivo:** Entender el schema final sin LLM.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Primero escribir JSON manual, luego validar.

```python
from pydantic import BaseModel
from typing import List

class ContractChangeOutput(BaseModel):
    sections_changed: List[str]
    topics_touched: List[str]
    summary_of_the_change: str

manual_json = {
    "sections_changed": ["duration"],
    "topics_touched": ["plazo contractual"],
    "summary_of_the_change": "La duración del contrato cambia de 12 a 18 meses."
}

ContractChangeOutput(**manual_json)
```
