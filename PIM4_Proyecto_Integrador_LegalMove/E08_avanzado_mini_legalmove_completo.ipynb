```python
from pydantic import BaseModel
from typing import List
import time

class ContractChangeOutput(BaseModel):
    sections_changed: List[str]
    topics_touched: List[str]
    summary_of_the_change: str

def log_span(name, input_preview, output_preview, latency_ms):
    return {
        "span_name": name,
        "input_preview": input_preview[:80],
        "output_preview": str(output_preview)[:120],
        "latency_ms": latency_ms
    }

def parse_contract_image(path):
    return "Contrato: pago mensual de $1000. Duración: 12 meses."

def contextualization_agent(original, amendment):
    return "Mapa: secciones payment_terms y duration detectadas."

def extraction_agent(original, amendment, context):
    return {
        "sections_changed": ["payment_terms"],
        "topics_touched": ["monto mensual"],
        "summary_of_the_change": "El monto mensual cambia de $1000 a $1200."
    }

trace = []

start = time.time()
original = parse_contract_image("original.png")
trace.append(log_span("parse_original_contract", "original.png", original, int((time.time()-start)*1000)))

start = time.time()
amendment = "Adenda: pago mensual de $1200. Duración: 12 meses."
trace.append(log_span("parse_amendment_contract", "amendment.png", amendment, int((time.time()-start)*1000)))

start = time.time()
context = contextualization_agent(original, amendment)
trace.append(log_span("contextualization_agent", original + amendment, context, int((time.time()-start)*1000)))

start = time.time()
raw = extraction_agent(original, amendment, context)
validated = ContractChangeOutput(**raw)
trace.append(log_span("extraction_agent", context, validated.model_dump(), int((time.time()-start)*1000)))

print(validated)
trace
```
