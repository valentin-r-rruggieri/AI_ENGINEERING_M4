# AEM4L3 E02 — Tool ecommerce con schema

**Objetivo:** Diseñar contrato de tool.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Toda tool debe tener input, output, versión y scope.

```python
import json

tool_contract = {
    "tool": "create_return_request",
    "version": "1.0",
    "input_schema": {"order_id": "string", "reason": "string", "customer_email": "string"},
    "output_schema": {"return_id": "string", "status": "created"},
    "required_scope": "returns:write"
}

print(json.dumps(tool_contract, indent=2, ensure_ascii=False))
```
