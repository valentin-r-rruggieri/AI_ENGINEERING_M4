# AEM4L3 E04 — Elegir transporte

**Objetivo:** Decidir STDIO vs HTTP streaming.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

El transporte define límites de seguridad, auditoría y escalamiento.

```python
scenarios = {
    "Plugin local de VS Code": ("STDIO", "Local, simple, baja latencia"),
    "Copilot interno de soporte": ("HTTP streaming", "Multiusuario y auditable"),
    "SaaS multi-tenant": ("HTTP streaming", "Seguridad, tenants y escala"),
    "Script local dev-only": ("STDIO", "Un usuario y bajo riesgo")
}

for s, (transport, reason) in scenarios.items():
    print(f"{s}: {transport} — {reason}")
```
