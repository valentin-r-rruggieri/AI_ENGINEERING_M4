# AEM4L3 E01 — Clasificar primitivas MCP

**Objetivo:** Distinguir tools, resources y prompts.

**Tipo:** Resuelto

**Uso:** listo para Google Colab. Ejecutá las celdas en orden.

Las tools hacen acciones; resources aportan contexto; prompts estandarizan instrucciones.

```python
capabilities = {
    "Crear ticket": "Tool",
    "Leer política de reembolso": "Resource",
    "Template de respuesta amable": "Prompt",
    "Enviar email": "Tool",
    "Consultar últimas compras": "Resource",
    "Actualizar estado del reclamo": "Tool",
    "Guía de tono institucional": "Prompt"
}

for name, primitive in capabilities.items():
    print(f"{name:35} → {primitive}")
```
